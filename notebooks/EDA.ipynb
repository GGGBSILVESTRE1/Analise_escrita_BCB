```python
# Importação de bibliotecas que serão utilizadas para a análise exploratória dos dados
import os
import sys
import pandas as pd 
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

sys.path.append(os.path.abspath(os.path.join('..')))
# funções de formatação e processamento de texto
from src.formatting import data_formatada
from src.processing import limpar_texto_para_tokenizar, tokenizar_texto, limpeza_para_nuvem
from src.analise_texto import hawkish_dovish

BASE_PATH = os.path.abspath(os.path.join(os.path.dirname("__file__"), ".."))
```

```python
# Carregando o DataFrame a partir do arquivo parquet
df = pd.read_parquet(os.path.join(BASE_PATH , "data/raw/atas.parquet"))


```

```python
#visualização das primeiras linhas do DataFrame

print(df.head())

df['DataReferencia'] = df['DataReferencia'].apply(data_formatada)
```

```python
# Informação sobre as colunas do DataFrame
df.info()

print(df['DataReferencia'])
```

```python
#visualizando o contéudo de uma ata para verificar se a extração foi bem sucedida
amostra = df['conteudo_ata'].iloc[2]

print(f"Tipo do dado: {type(amostra)}")
print(f"Quantidade de caracteres: {len(amostra) if amostra else 0}")
print("-" * 30)
print(amostra[:1000] if amostra else "NADA FOI EXTRAÍDO")
```

```python
# Limpeza e tokenização
df['texto_limpo'] = df['conteudo_ata'].apply(limpar_texto_para_tokenizar)

df['tokens'] = df['texto_limpo'].apply(tokenizar_texto)

df['tokens_nuvem'] = df['tokens'].apply(limpeza_para_nuvem)

#Visualização da nuvem de palavras 
freq = Counter(sum(df['tokens_nuvem'], [])) # contagem de frequência de palavras
wordcloud = WordCloud(width = 800, height = 400 , background_color='white', colormap='viridis').generate_from_frequencies(freq)

plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Análise de hawkish(mais agressivo) e dovish(mais brando)
df['score'] = df['tokens'].apply(hawkish_dovish)

df['hawkish_or_dovish'] =  df['score'].apply(lambda x: 'Hawkish' if x > 0 else ('Dovish' if x < 0 else 'Neutro'))

df['hawkish_or_dovish'].value_counts()




```
